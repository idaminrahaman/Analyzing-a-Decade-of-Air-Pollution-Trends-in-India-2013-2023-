```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
df = pd.read_csv('C:\\DATA ANALYSIS\\My projects\\polution-data\\12_final_cleaned_file.csv')
df.drop(['Unnamed: 0', 'id'], axis=1, inplace=True)
df['state_ut'] = df['state_ut'].str.capitalize()
df['location'] = df['location'].str.capitalize()
```

```python
#Region wise aqi distribution

region_aqi = (
    df.groupby(['region_zone', 'Year'])['AQI']
    .mean()
    .reset_index(name='Average_AQI')
)

plt.figure(figsize=(10, 6))

for region in region_aqi['region_zone'].unique():
    subset = region_aqi[region_aqi['region_zone'] == region]
    plt.plot(subset['Year'], subset['Average_AQI'], marker='o', label=region)

plt.xlabel('Year')
plt.ylabel('Average AQI')
plt.title('Region-wise Average AQI Over Time')
plt.legend(title='Region')
plt.grid(True)

plt.xticks(sorted(df['Year'].unique()))

plt.tight_layout()
plt.show()
print(region_aqi[region_aqi['region_zone'] == 'North'])
```

```python
#Region wise so2 distribution
# Group by region and year, then compute mean SO₂
region_data = (
    df.groupby(['region_zone', 'Year'])['so2']
    .mean()
    .reset_index(name='Average_so2')
)

# Create the plot
plt.figure(figsize=(10, 6))  # optional, for better readability

# Loop through each region and plot a line
for region in region_data['region_zone'].unique():
    subset = region_data[region_data['region_zone'] == region]
    plt.plot(subset['Year'], subset['Average_so2'], marker='o', label=region)

# Add labels, title, legend, and grid
plt.xlabel('Year')
plt.ylabel('Average SO₂ Level')
plt.title('Average SO₂ Levels by Region Over Time')
plt.legend(title='Region')
plt.grid(True)

# Show all year ticks on the x-axis
plt.xticks(region_data['Year'].unique())
plt.show()
```
